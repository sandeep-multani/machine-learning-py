```python
import pandas

dataframe = pandas.read_csv('https://media.githubusercontent.com/media/sandeep-multani/datasets/main/datasets/loan_data.csv')

dataframe.head(5)
```

```python
dataframe.describe()
```

```python
dataframe.info()
```

```python
#dataframe = dataframe.dropna()
dataframe.dropna(inplace=True)
```

```python
dataframe['Gender'].replace({'Male':1, 'Female':0}, inplace=True)
dataframe['Education'].replace({'Graduate':1, 'Not Graduate':0}, inplace=True)
```

```python
dataframe['Gender'].unique()
```

```python
dataframe['Education'].unique()
```

```python
print(dataframe['Married'].unique())
dataframe['Married'].replace({'Yes':1, 'No':0}, inplace=True)
print(dataframe['Married'].unique())
```

```python
print(dataframe['Dependents'].unique())

for x in dataframe.index:
  if dataframe.loc[x, "Dependents"] == "3+":
    dataframe.loc[x, "Dependents"] = "4"

print(dataframe['Dependents'].unique())    
```

```python
dataframe['Dependents'] = pandas.to_numeric(dataframe['Dependents'])
print(dataframe['Dependents'].unique()) 
```

```python
print(dataframe['Self_Employed'].unique())
print(dataframe['Loan_Status'].unique())

dataframe['Self_Employed'].replace({'Yes':1, 'No':0}, inplace=True)
dataframe['Loan_Status'].replace({'Y':1, 'N':0}, inplace=True)

print(dataframe['Self_Employed'].unique())
print(dataframe['Loan_Status'].unique())
```

```python
print(dataframe['Property_Area'].unique())
dataframe['Property_Area'].replace({'Rural':0, 'Urban':1, 'Semiurban':2}, inplace=True)
print(dataframe['Property_Area'].unique())
```

```python
del dataframe['Loan_ID']
dataframe.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.heatmap(dataframe.corr(),
			cmap = 'BrBG',
			fmt = '.2f',
			linewidths = 2,
			annot = True)
```

```python
X = dataframe.drop(['Loan_Status'], axis=1)
Y = dataframe['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split

X = dataframe.drop(['Loan_Status'],axis=1)
Y = dataframe['Loan_Status']
X.shape,Y.shape

X_train, X_test, Y_train, Y_test = train_test_split(X, Y,
													test_size=0.4,
													random_state=1)
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

from sklearn import metrics

knn = KNeighborsClassifier(n_neighbors=3)
rfc = RandomForestClassifier(n_estimators = 7, criterion = 'entropy', random_state =7)
svc = SVC()
lc = LogisticRegression()

for clf in (rfc, knn, svc, lc):
	clf.fit(X_train, Y_train)
	Y_pred = clf.predict(X_train)
	print("Accuracy score of ",
		clf.__class__.__name__,
		"=",100*metrics.accuracy_score(Y_train, Y_pred))
```

```python
for clf in (rfc, knn, svc, lc):
	clf.fit(X_train, Y_train)
	Y_pred = clf.predict(X_test)
	print("Accuracy score of ",
		clf.__class__.__name__,"=",
		100*metrics.accuracy_score(Y_test, Y_pred))
```
